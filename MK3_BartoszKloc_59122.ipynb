```python
import numpy as np
import pandas as pd

df = pd.read_csv("war2.csv", sep=";")
df["y"] = df["y"].astype(str).str.replace(",", ".", regex=False).astype(float)

A = np.column_stack([df["x1"].to_numpy(float), df["x2"].to_numpy(float)])
y = df["y"].to_numpy(float)

a, b = (np.linalg.pinv(A) @ y)
print("a =", a)
print("b =", b)
print("rank(A) =", np.linalg.matrix_rank(A))
print("a + 2b =", a + 2*b)
```
