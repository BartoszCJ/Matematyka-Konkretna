```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-6, 6, 2000)


y = np.tanh(x)


dy = 1 - y**2

plt.figure()
plt.plot(x, y, label="tanh(x)")
plt.plot(x, dy, label="gradient tanh(x)")
plt.title("Funkcja tanh oraz jej gradient")
plt.xlabel("x")
plt.grid(True)
plt.legend()
plt.show()
```

```python
eps = 1e-5
y_plus = np.tanh(x + eps)
y_minus = np.tanh(x - eps)
dy_num = (y_plus - y_minus) / (2*eps)

print("max |dy - dy_num| =", np.max(np.abs(dy - dy_num)))
```
